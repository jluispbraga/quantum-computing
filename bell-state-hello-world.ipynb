```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
```

#### 1. Criando um circuito com 2 qubits e 2 bit clássicos

```python
qc = QuantumCircuit(2, 2)
```

#### 2.Aplicando uma porta Hadmard (H) no qubit 0 para criar superposição
- Isso coloca o qubit 0 em superposição (|0> + |1>) / sqrt(2)

```python
qc.h(0)
```

#### 3.Aplica uma porta CNOT (Cx) para emaranhar o qubit 0 (controle) com o qubit 1 (alvo)
- Control=qubit 0, Target=qubit 1. Isso gera emaranhamento.

```python
qc.cx(0, 1)
```

#### 4.Realiza a medição dos qubits para os bits clássicos correspondentes
- Mapeia o qubit 0 no bit 0, e o qubit 1 no bit 1

```python
qc.measure([0, 1], [0, 1])
```

#### 5. Define o backend e executa

```python
backend = Aer.get_backend("qasm_simulator")
tqc = transpile(qc, backend)
job = backend.run(tqc, shots=1024) # Rodando o circuto 1024 vezes
```

#### 6.Resultados

```python
result = job.result()
counts = result.get_counts()
print(f"Resultados (Contagens): {counts}")
```

##### O que o resultado quer dizer?
Nesse teste o circuito foi executado 1024 vezes
- '00': 497 -> Quase metade dos experimentos ambos os qubits colapsaram para o estado 0
- '11': 527 -> Na outra metade, ambos colapsaram para o estado 1
- Nenhum dos qubist foram encontrados em estados diferentes

Conclusão -> Os qubits estão correlacionados de forma que, se você medir um e descobrir que ele é 1, o outro será obrigatoriamente 1 também, instantanemente, não importa a distância entre eles

#### Circuito quântico desenhado

```python
print(qc.draw(output='text'))
```

- Porta H (Hadamard)
    - Transforma um estado definino em um estado de superposição, onde o qubit existe como uma combinação linear de ambos os estados simultaneamente
- Porta X (Pauli-X) ou Porta NOT
    - Realiza uma inversão de bit (bit flip)
- M (Medição/Measurement)
    - É o processo de observar o qubit para extrair informação quântica e convertê-la em bits clássicos (0 ou 1)
